```python
def categorical_plot(features, df, segment_feature=None):
    fig, ax = plt.subplots(len(features), 1,figsize=(8,35))
    for i, feature in enumerate(features):
        if segment_feature:
            sns.countplot(data=df, x=segment_feature, hue=feature, ax=ax[i])
        else:
            sns.countplot(data=df, x=feature, ax=ax[i])
    plt.tight_layout()
    plt.show()
```

```python
def numerical_dis_plot(features, df, segment_feature=None, showfliers=True):
    fig, ax = plt.subplots(len(features), 1,figsize=(8,30))
    for i, feature in enumerate(features):
        if segment_feature:
            sns.boxplot(y=segment_feature, x=feature, data=df, ax=ax[i], showfliers=showfliers)
            ax[i].set_ylabel(None)
        else:
            sns.boxplot(x=feature, data=df, ax=ax[i], showfliers=showfliers)
    plt.tight_layout()
    plt.show()
```

# RFM Analysis

* Recency: Seberapa baru karyawan terakhir kali mendapatkan promosi
* Frequency: Seberapa sering karyawan mendapatkan pelatihan 
* Monetary: Total kompensasi yang diterima karyawan setiap bulan

```python
employee_df['Recency'] = employee_df['YearsSinceLastPromotion']
employee_df['Recency'] = employee_df['Recency'].max() - employee_df['Recency']
```

```python
employee_df['Frequency'] = employee_df['TrainingTimesLastYear']
```

```python
employee_df['Monetary'] = employee_df['MonthlyIncome']
```

```python
def assign_rfm_scores(series, ascending=True):
    if ascending:
        return pd.qcut(series.rank(method='first'), 4, labels=[4, 3, 2, 1])
    else:
        return pd.qcut(series.rank(method='first'), 4, labels=[1, 2, 3, 4])
```

```python
employee_df['R_Score'] = assign_rfm_scores(employee_df['Recency'], ascending=False)
employee_df['F_Score'] = assign_rfm_scores(employee_df['Frequency'])
employee_df['M_Score'] = assign_rfm_scores(employee_df['Monetary'])
```

```python
employee_df['R_Score'] = employee_df['R_Score'].astype(int)
employee_df['F_Score'] = employee_df['F_Score'].astype(int)
employee_df['M_Score'] = employee_df['M_Score'].astype(int)
```

```python
employee_df['R_Score_Norm'] = (employee_df['R_Score']/employee_df['R_Score'].max())*100
employee_df['F_Score_Norm'] = (employee_df['F_Score']/employee_df['F_Score'].max())*100
employee_df['M_Score_Norm'] = (employee_df['M_Score']/employee_df['M_Score'].max())*100
```

```python
employee_df['RFMScore'] = (0.2*employee_df['R_Score_Norm'] + 0.3*employee_df['F_Score_Norm'] + 0.5*employee_df['M_Score_Norm']) * 0.05
```

```python
rfm_df = employee_df[['EmployeeId', 'Recency', 'Frequency', 'Monetary', 'R_Score', 'F_Score', 'M_Score', 'R_Score_Norm', 'F_Score_Norm', 'M_Score_Norm', 'RFMScore']]
rfm_df.head()
```

```python
def rfm_score(employee_df):
    employee_df = employee_df.copy()

    # Calculate the rank
    employee_df['R_Score'] = employee_df['Recency'].rank(ascending=False)
    employee_df['F_Score'] = employee_df['Frequency'].rank(ascending=True)
    employee_df['M_Score'] = employee_df['Monetary'].rank(ascending=True)

    # normalizing the rank of the customers
    employee_df['R_Score_Norm'] = (employee_df['R_score']/employee_df['R_score'].max())*100
    employee_df['F_Score_Norm'] = (employee_df['F_score']/employee_df['F_score'].max())*100
    employee_df['M_Score_Norm'] = (employee_df['M_score']/employee_df['M_score'].max())*100

    # Calculate the RFM score
    employee_df['RFMScore'] = (0.2*employee_df['R_Score_Norm'] + 0.3*employee_df['F_Score_Norm'] + \
        0.5*employee_df['M_Score_Norm']) * 0.05
    employee_df = employee_df.round(2)

    return employee_df
```
